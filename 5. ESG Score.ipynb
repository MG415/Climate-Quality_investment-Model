```python
stock_list = ['AAPL',  'APTV', 'BAC', 'BALL',  'BEP', 'BKNG', 'BMRN', 'COST', 'CVS', 'DE', 'ETN', 'GOOGL', 'HD', 'HIG',  'MS', 'MSFT', 'PLD',  'TT', 'UNH', 'V']
```

```python
df = pd.DataFrame()
for i in stock_list:
    df[i+'_Total-Score'] = yesg.get_historic_esg(i)['Total-Score']
    df[i+'_E-Score'] = yesg.get_historic_esg(i)['E-Score']
    df[i+'_S-Score'] = yesg.get_historic_esg(i)['S-Score']
    df[i+'_G-Score'] = yesg.get_historic_esg(i)['G-Score']
```
