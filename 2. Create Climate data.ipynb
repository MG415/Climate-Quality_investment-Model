# 4. Download all climate data

### 4.1 Download all climate data

```python
cdd_18  = pd.read_csv('CDD_18.csv', index_col='Country')
cdd_21  = pd.read_csv('CDD_21.csv', index_col='Country')
irradiance  = pd.read_csv('Global_horizontal_irradiance.csv', index_col='Country')
hdd_16  = pd.read_csv('HDD_16.csv', index_col='Country')
hdd_18  = pd.read_csv('HDD_18.csv', index_col='Country')
heat  = pd.read_csv('Heat_index.csv', index_col='Country')
humidity  = pd.read_csv('Relative_humidity.csv', index_col='Country')
temperature = pd.read_csv('Temperature.csv', index_col='Country')
precipitation  = pd.read_csv('Total_precipitation.csv', index_col='Country')
```

### 4.2 Change column name for the top 8 industrial countries

```python
for i in cdd_18.columns:
    cdd_18.rename(columns = {i:'cdd_18-'+i}, inplace = True)

cdd_18.head(1)
```

```python
for i in cdd_21.columns:
    cdd_21.rename(columns = {i:'cdd_21-'+i}, inplace = True)

cdd_21.head(1)
```

```python
for i in irradiance.columns:
    irradiance.rename(columns = {i:'irradiance-'+i}, inplace = True)

irradiance.head(1)
```

```python
for i in hdd_16.columns:
    hdd_16.rename(columns = {i:'hdd_16-'+i}, inplace = True)

hdd_16.head(1)
```

```python
for i in hdd_18.columns:
    hdd_18.rename(columns = {i:'hdd_18-'+i}, inplace = True)

hdd_18.head(1)
```

```python
for i in heat.columns:
    heat.rename(columns = {i:'heat-'+i}, inplace = True)

heat.head(1)
```

```python
for i in humidity.columns:
    humidity.rename(columns = {i:'humidity-'+i}, inplace = True)

humidity.head(1)
```

```python
for i in temperature.columns:
    temperature.rename(columns = {i:'temperature-'+i}, inplace = True)

temperature.head(1)
```

```python
for i in precipitation.columns:
    precipitation.rename(columns = {i:'precipitation-'+i}, inplace = True)

precipitation.head(1)
```

# 5. Combine all climate data in one dataframe

```python
climate = pd.concat([cdd_18,
                     cdd_21.reindex(cdd_18.index),
                    irradiance.reindex(cdd_18.index),
                    hdd_16.reindex(cdd_18.index),
                    hdd_18.reindex(cdd_18.index),
                    heat.reindex(cdd_18.index),
                    humidity.reindex(cdd_18.index),
                    temperature.reindex(cdd_18.index),
                    precipitation.reindex(cdd_18.index)], axis=1)
```

```python
climate
```
